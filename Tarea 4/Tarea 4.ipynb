#Punto 1
hallar la serie de Fourier para la función que es $x^2$ para $0 < x < \pi $ y 0 para $\pi < x < 2\pi $. La serie resulta ser:
$$
f(x)=\frac{\pi^2}{6}+\sum_{n=1}^{\infty}\left[\frac{2(-1)^n}{n^2}cos(nx)+\frac{2(-1)^n-(-1)^n(n\pi)^2-2}{n^3\pi}sin(nx) \right]
$$

```python
points = 1000
x=linspace(0,2*pi,points)
def f(x):
    if (x>= 0 and x< pi):
        return x**2
    else:
        return 0
```

```python
func=[]
for i in range(len(x)):
    func.append(f(x[i]))
```

```python
def C(n):
    return (-1)**n*2/(n+0.0)**2
def S(n):
    return (2*(-1)**n-(-1)**n*(n*pi)**2-2)/(n**3*pi)
```

```python
def fourier(n,x):
    ans = pi**2/6
    for i in range(n):
        ans += C(i+1)*cos((i+1)*x)+S(i+1)*sin((i+1)*x)
    return ans
```

```python
func_5=fourier(5,x)
func_15=fourier(15,x)
func_30=fourier(30,x)
plot(x,func)
plot(x,func_5)
plot(x,func_15)
plot(x,func_30)

legend(['$f(x)$', '5 terms', '15 terms', '30 terms'])
ax = axes()
ax.set_xlabel("$x$",fontsize=20)
ax.set_ylabel("$y$",fontsize=20)
ax.set_title("Function and fourier series 5,15 and 30 terms")
filename = 'figura punto 1'
savefig(filename + '.pdf',format = 'pdf', transparent=True)
```

#Punto 3
Se tiene una masa que se mueve armónicamente y está sometida a una fuerza $f(t)=F_0\phi(t)$ donde $\phi(t)$ es 1 entre 0 y 0.5 y -1 entre 0.5 y 1, y continua periodicamente. A continuación se muestra la gráfica de posición vs tiempo para esta configuración.
La solución resulta ser:
$$
x=\sum_{n \text{ odd}}\xi_{n}cos(\omega_n t - \delta_n) \\
\xi_n = \frac{A_n}{m\omega_0^2}\left( \left(1-\frac{\omega_n^2}{\omega_0^2}\right)^2 + \frac{\omega_n^2}{Q^2\omega_0^2}\right) \\
\delta_k=arctan\left(\frac{\frac{\omega_n}{Q\omega_0}}{\frac{\omega_n^2}{\omega_0^2}-1}\right)
$$
Donde $A_n$ son os coeficientes de fourier de la fuerza
$$
F(x) = \sum_{n \text{ odd}} A_n sin(\omega_n t)\\
A_n = \frac{4F_0}{n\pi}
$$

```python
F0=500
k=10**3
m=1
Q=100
w0=sqrt(k/m)
def w(n):
    return 2*n*pi
def A(n):
    return 4.0*F0/(pi*n)
def xi(n):
    return A(n)/(m*w0**2.0)/sqrt((1.0-(w(n)**2/w0**2))**2+ w(n)**2/(Q**2*w0**2))
def d(n):
    return math.atan(((w(n))/(w0*Q))/(w(n)**2/(w0)**2-1.0))
```

```python
def f(x):
    ans=0
    for n in range(1000):
        ans+= A(2*n+1)*sin(w(2*n+1)*x)
    return ans
```

```python
x= linspace(0,2,1000)
plot(x,f(x))
```

```python
npoints=10000
N=1000
T=2
pos = zeros(npoints)
t=linspace(0,T,npoints)
def pos(t):
    ans = 0
    for n in range(N):
        ans+=xi(2*n+1)*cos(w(2*n+1)*t-d(2*n+1))
    return ans
```

```python
plot (t,pos(t))
ax = axes()
ax.set_xlabel("$t$",fontsize=20)
ax.set_ylabel("$x$",fontsize=20)
ax.set_title("Position against time")
filename = 'figura punto 3'
savefig(filename + '.pdf',format = 'pdf', transparent=True)
```
