#Punto 1
hallar la serie de Fourier para la función que es $x^2$ para $0 < x < \pi $ y 0 para $\pi < x < 2\pi $. La serie resulta ser:
$$
f(x)=\frac{\pi^2}{6}+\sum_{n=1}^{\infty}\left[\frac{2(-1)^n}{n^2}cos(nx)+\frac{2(-1)^n-(-1)^n(n\pi)^2-2}{n^3\pi}sin(nx) \right]
$$

```python
points = 1000
x=linspace(0,2*pi,points)
def f(x):
    if (x>= 0 and x< pi):
        return x**2
    else:
        return 0
```

```python
func=[]
for i in range(len(x)):
    func.append(f(x[i]))
```

```python
def C(n):
    return (-1)**n*2/(n+0.0)**2
def S(n):
    return (2*(-1)**n-(-1)**n*(n*pi)**2-2)/(n**3*pi)
```

```python
def fourier(n,x):
    ans = pi**2/6
    for i in range(n):
        ans += C(i+1)*cos((i+1)*x)+S(i+1)*sin((i+1)*x)
    return ans
```

```python
func_5=fourier(5,x)
func_15=fourier(15,x)
func_30=fourier(30,x)
plot(x,func)
plot(x,func_5)
plot(x,func_15)
plot(x,func_30)

legend(['$f(x)$', '5 terms', '15 terms', '30 terms'])
ax = axes()
ax.set_xlabel("$x$",fontsize=20)
ax.set_ylabel("$y$",fontsize=20)
ax.set_title("Function and fourier series 5,15 and 30 terms")
filename = 'figura punto 1'
savefig(filename + '.pdf',format = 'pdf', transparent=True)
```
