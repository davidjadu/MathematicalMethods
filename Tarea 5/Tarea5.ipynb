```python
%pylab inline
from cmath import phase
```

```python
C=10**(-9) #Farads
L=10**(-3) #Henries
R=10**3 #Ohms
```

```python
def z(w):
    return R-(1j*w*L)/(1-C*L*w**2)
```

```python
w=linspace(0,2E6,1000)
```

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))

ax.plot(w, abs(1.0/z(w)),color='r')
ax.set_title("Transfer function")
ax.set_xlabel("$\omega[rad/s]$",fontsize=16)
ax.set_ylabel("$|1/z(\omega)|$",fontsize=16)
filename = 'Función de transferencia'
savefig(filename + '.pdf',format = 'pdf', transparent=True)
```

```python
phase_array = 1/z(w)
for i in range(len(phase_array)):
    phase_array[i]=phase(phase_array[i])
```

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))

ax.plot(w, phase_array,color='b')
ax.set_title("Transfer function phase")
ax.set_xlabel("$\omega[rad/s] $",fontsize=16)
ax.set_ylabel("$Arg(1/z(\omega))[rad]$",fontsize=16)
filename = 'Fase función transferencia'
savefig(filename + '.pdf',format = 'pdf', transparent=True)
```
